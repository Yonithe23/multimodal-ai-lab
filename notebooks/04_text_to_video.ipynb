```python
import os
import time
from dotenv import load_dotenv
from openai import OpenAI
from IPython.display import Video, display
```

```python
load_dotenv()
client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))
```

```python
prompt = """
Create a short realistic video of a student learning AI on a laptop
in a modern classroom with soft lighting.
"""
```

```python
video = client.videos.create(
    model="sora-2",
    prompt=prompt,
    size="1280x720",
    seconds="4"
)
```

```python
while video.status not in ["completed", "failed"]:
    time.sleep(5)
    video = client.videos.retrieve(video.id)
    print("Status:", video.status)
```

```python
response = client.videos.download_content(video.id)
content = response.read()

with open("text_to_video3.mp4", "wb") as f:
    f.write(content)
```

```python
display(Video("text_to_video3.mp4", embed=True))
```
