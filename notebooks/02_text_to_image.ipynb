```python
import os
import base64
from dotenv import load_dotenv
from openai import OpenAI
from PIL import Image
from IPython.display import display
```

```python
# Load API key from .env file

load_dotenv()

if os.getenv("OPENAI_API_KEY"):
    print(f"✅ OPENAI_API_KEY found , {os.getenv("OPENAI_API_KEY")[:10]}")
else:
    print("❌ OPENAI_API_KEY not found")
    print("   Create a .env file with: OPENAI_API_KEY=your-key-here")

# Check for Google API key (optional, for Google Gemini)
if os.getenv("GOOGLE_API_KEY"):
    print(f"✅ GOOGLE_API_KEY found , {os.getenv("GOOGLE_API_KEY")[:10]}")
else:
    print("⚠️  GOOGLE_API_KEY not found (optional for this notebook)")


OPENAI_API_KEY = os.getenv("OPENAI_API_KEY")

print("API Key Loaded Successfully")
print(f"The api key is , {OPENAI_API_KEY [:10]}")

client = OpenAI(
    api_key=os.getenv("OPENAI_API_KEY")
)

print("API Key Loaded Successfully")
```

```python
# Text prompt

prompt = """
Create a modern AI classroom image.
A student is using a laptop, with glowing AI icons around the screen.
Style: realistic, clean, professional.
"""
```

```python
result = client.images.generate(
    model="gpt-image-1",
    prompt=prompt,
    size="1024x1024"
)
```

```python
image_base64 = result.data[0].b64_json
image_bytes = base64.b64decode(image_base64)

with open("text_to_image.png", "wb") as f:
    f.write(image_bytes)

print("Image saved successfully")
```

```python
image = Image.open("text_to_image.png")
display(image)
```
