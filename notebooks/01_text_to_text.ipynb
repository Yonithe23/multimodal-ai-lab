```python
# Step 3: Load environment variables (API keys)
from dotenv import load_dotenv
import os

# Load .env file
load_dotenv()

# Check if OpenAI API key is set
# NOTE: We don't print the actual key for security!

if os.getenv("OPENAI_API_KEY"):
    print(f"✅ OPENAI_API_KEY found , {os.getenv("OPENAI_API_KEY")[:10]}")
else:
    print("❌ OPENAI_API_KEY not found")
    print("   Create a .env file with: OPENAI_API_KEY=your-key-here")

# Check for Google API key (optional, for Google Gemini)
if os.getenv("GOOGLE_API_KEY"):
    print(f"✅ GOOGLE_API_KEY found , {os.getenv("GOOGLE_API_KEY")[:10]}")
else:
    print("⚠️  GOOGLE_API_KEY not found (optional for this notebook)")



OPENAI_API_KEY = os.getenv("OPENAI_API_KEY")

print("API Key Loaded Successfully")
print(f"The api key is , {OPENAI_API_KEY }")
```

```python
# =========================================================
# Create LLM Model
# =========================================================
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(
    model="gpt-4.1-mini",
    api_key=OPENAI_API_KEY,
    temperature=0.7
)
```

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
```

```python
# =========================================================
# Create Prompt Template
# =========================================================

prompt = ChatPromptTemplate.from_messages(
    [
        ("system", "You are a helpful AI assistant."),
        ("human", "{question}")
    ]
)

# =========================================================
# Output Parser
# =========================================================


output_parser = StrOutputParser() 

# =========================================================
# Create LangChain Pipeline
# =========================================================

chain = prompt | llm | output_parser

# =========================================================
# User Input
# =========================================================

user_question = "Explain what Machine Learning is in simple words."

# =========================================================
# Generate Response
# =========================================================

response = chain.invoke(
    {
        "question": user_question
    }
)

print(response)
```

```python
# =========================================================
# Interactive Chat
# =========================================================

while True:
    
    question = input("You: ")
    
    if question.lower() == "exit":
        print("Chat Ended")
        break

    response = chain.invoke(
        {
            "question": question
        }
    )

    print("\nAI:", response)
    print()
```
